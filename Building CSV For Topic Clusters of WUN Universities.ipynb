```python
import powerlaw #fit field weighted citation impact to a power law and produce 5% conf critical value
```

```python
results=powerlaw.Fit(df['Field-Weighted Citation Impact'])
```

```python
alpha=results.power_law.alpha
xmin=results.power_law.xmin
R, p = results.distribution_compare('power_law', 'lognormal')
```

```python
def power(x): 
    return (alpha-1)/(xmin)*(x/2.6)**-2.92
```

```python
import numpy as np
import scipy.integrate
```

```python
xs=np.linspace(2.6,1000)
```

```python
for x in xs:
    val=scipy.integrate.quad(power,x,np.inf)[0]
    if val <=.05:
        print(x)
        break
#seems high
```
