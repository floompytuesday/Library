```python
import powerlaw
df.dropna()
results=powerlaw.Fit(df['Field-Weighted Citation Impact'])
alpha=results.alpha
R, p = results.distribution_compare('power_law', 'lognormal')
print(R,p)
R,p=results.distribution_compare('power_law','exponential')
print(R,p)
ks=results.D
```

```python
xmin=results.xmin
import numpy as np
import scipy.stats 
def power(x): 
    return ((alpha-1)/(xmin))*((x/xmin)**(-1*alpha))
scipy.integrate.quad(power,xmin,np.inf)[0]
xmin
```

```python
xs=np.linspace(1.63,20,1000)
for fw in xs:
    val=scipy.integrate.quad(power,fw,np.inf)[0]
    if val <=.05:
        print(fw)
        break
```
