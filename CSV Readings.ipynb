```python
import pandas as pd
import os
def parse_csv(file,top_50=True):
    dat=pd.read_csv(file,skiprows=12)
    new=dat.sort_values('Field-Weighted Citation Impact',ascending=False)[0:50]
    if top_50:
        return new
    else:
        return dat
```

```python
dfs={}
for filename in set(os.listdir()):
    if filename.endswith('.csv'):
        dfs[filename]=parse_csv(filename) #access df with name of file as key
world=parse_csv('Top_500_Topic_Clusters_by_Scholarly_Output-World.csv',top_50=False)
```
